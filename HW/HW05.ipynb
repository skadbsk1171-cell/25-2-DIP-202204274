### Exercise 4.
- highpass filtering   blurred image sharp  .
-  blurred image , Laplacian kernel  highpass filtering     sharpness .
- w  c     .

```python
f = io.imread(' img01.jpeg',as_gray=True)[::2,::2]

w = 0.5 * np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
y = signal.convolve2d(f,w,mode='same')
c = 0.03
g = f + c*(y)
plt.figure(figsize=(18,6))
plt.subplot(131)
plt.imshow(f,cmap='gray')
plt.colorbar()
plt.axis('off')
plt.title('original')
plt.subplot(132)
plt.imshow(g,cmap='gray')
plt.colorbar()
plt.axis('off')
plt.title('sharpening')
plt.subplot(133)
plt.imshow(y,cmap='RdBu_r',vmin=-0.5*np.max(np.abs(y)),vmax=0.5*np.max(np.abs(y)))
plt.colorbar()
plt.axis('off')
plt.title('filtered')
plt.show()
```
